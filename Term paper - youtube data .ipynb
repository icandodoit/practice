유튜브(YouTube)는 사용자가 동영상을 자유롭게 올리거나 볼 수 있는 구글의 콘텐츠 호스팅 웹사이트이다. 명실상부 세계 최대 규모의 비디오 플랫폼이다.

명칭은 사용자를 가리키는 '유'(You, 당신)와 미국 영어에서 텔레비전의 별칭인 '튜브(tube)'의 합성이다. 과거 텔레비전이 브라운관(Cathode-Ray Tube, CRT)를 사용했기 때문에 텔레비전을 미국 영어에서 다른 말로 '튜브'라 부른다.[4] 즉 '유튜브'라는 명칭의 뉘앙스는 '당신을 위한 텔레비전', '당신이 곧 텔레비전' 정도이다. 네모난 유튜브 아이콘의 끝부분은 둥글게 깎여 있고 각 변은 끝부분보다 볼록하게 나와 있는데 옛 텔레비전에 쓰였던 브라운관의 시각상 특징을 아이콘으로 디자인한 것이다. 일명 '너' 튜브로도 통용된다.

# likes (좋아요 수) 

## likes 파일 전체 분석

```python
youtube_data.loc[youtube_data['likes'] >= 2000000, ['video_id', 'title'] ].sort_values(by=['title'], ascending=True).drop_duplicates()
```

```python
# dislikes의 평균은 GB(Great Britain)가 가장 높음
search = youtube_data.groupby('country').agg(
       {'dislikes': np.mean}).sort_values(by='dislikes',ascending=False)
search
```

```python
plt.scatter(youtube_data['likes'],youtube_data['dislikes'])
plt.xlabel("likes")
plt.ylabel("dislikes")
plt.show()
```

```python
youtube_data.loc[(youtube_data['likes'] >= 1000000) & (youtube_data['dislikes'] <= 5000), ['video_id', 'title','country']].drop_duplicates()
```

```python
youtube_data['like_ratio'] = youtube_data['likes'] / (youtube_data['likes'] + youtube_data['dislikes'])
youtube_data['like_ratio_views'] =  (youtube_data['likes'] + youtube_data['dislikes'])/youtube_data['views'] 
youtube_data['likes/views'] =  youtube_data['likes']/youtube_data['views']
youtube_data['dislikes/views'] =  youtube_data['dislikes']/youtube_data['views']
```

```python
# like_ratio = likes / (likes + dislikes)의 평균은 GB(Great Britain)가 가장 높음
search = youtube_data.groupby('country').agg(
       {'like_ratio': np.mean}).sort_values(by='like_ratio',ascending=False)
search
```

```python
# like_ratio = (likes + dislikes) / views 
search = youtube_data.groupby('country').agg(
       {'like_ratio_views': np.mean}).sort_values(by='like_ratio_views',ascending=False)
search
```

```python
# likes / views 
search = youtube_data.groupby('country').agg(
       {'likes/views': np.mean}).sort_values(by='likes/views',ascending=False)
search
```

```python
# dislikes / views 
search = youtube_data.groupby('country').agg(
       {'dislikes/views': np.mean}).sort_values(by='dislikes/views',ascending=False)
search
```
