```python
data_result_dropNaN = pd.merge(CCTV_Busan, pop_Busan, on='')
# NaN   
data_result_dropNaN.drop([15], inplace=True)
data_result_dropNaN
#index
data_result_dropNaN.set_index('', inplace=True)
```

```python
fp1 = np.polyfit(data_result_dropNaN[''],data_result_dropNaN[''], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(50000, 400000, 100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result_dropNaN[''], data_result_dropNaN[''], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

###      

```python
fp1 = np.polyfit(data_result_dropNaN[''],data_result_dropNaN[''], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(50000, 400000, 100)

data_result_dropNaN[''] = np.abs(data_result_dropNaN[''] - f1(data_result_dropNaN['']))

df_sort = data_result_dropNaN.sort_values(by='', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result_dropNaN[''], data_result_dropNaN[''], 
            c=data_result_dropNaN[''], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort[''][n]*1.02, df_sort[''][n]*0.98, 
             df_sort.index[n], fontsize=15)
    
plt.xlabel('')
plt.ylabel('')
plt.colorbar()
plt.grid()
plt.show()
```
