```python
data_result['CCTV비율'] = data_result['소계'] / data_result['인구'] * 100

data_result['CCTV비율'].sort_values().plot(kind='barh', 
                                         grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result_dropNaN = pd.merge(CCTV_Busan, pop_Busan, on='구별')
# NaN이 있는 행을 삭제함
data_result_dropNaN.drop([15], inplace=True)
data_result_dropNaN
#index
data_result_dropNaN.set_index('구별', inplace=True)
```

```python
fp1 = np.polyfit(data_result_dropNaN['인구'],data_result_dropNaN['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(50000, 400000, 100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result_dropNaN['인구'], data_result_dropNaN['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

### 조금 더 설득력 있는 자료 만들기

```python
fp1 = np.polyfit(data_result_dropNaN['인구'],data_result_dropNaN['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(50000, 400000, 100)

data_result_dropNaN['오차'] = np.abs(data_result_dropNaN['소계'] - f1(data_result_dropNaN['인구']))

df_sort = data_result_dropNaN.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result_dropNaN['인구'], data_result_dropNaN['소계'], 
            c=data_result_dropNaN['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구'][n]*1.02, df_sort['소계'][n]*0.98, 
             df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```

- 부산시에서 다른 구와 비교했을 때, 기장군, 사하구, 남구, 동구는 CCTV가 많음
- 연제구, 해운대구, 영도구, 서구, 중구, 동래구는 다른 구에 비해 CCTV 비율이 낮음
